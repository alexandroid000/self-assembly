This notebook takes the sparse and full hitting time algorithm results and compares the differences

```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
```

```python
full = pd.read_csv('../../data/full.csv', index_col=False, sep=',', header=None)
sparse = pd.read_csv('../../data/sparse.csv', index_col=False, sep=',', header=None)
```

```python
N = np.sqrt(full.shape[0])
N = int(N)
N
```

```python
full = full.values
sparse = sparse.values
```

```python
full = full.flatten()
sparse = sparse.flatten()
```

```python
full_2d = full.reshape(128,128)
sparse_2d = sparse.reshape(128,128)
```

```python
difference = full_2d - sparse_2d
sns.heatmap(difference)
```
